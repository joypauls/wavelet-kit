```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Morlet wavelet function
def morlet_wavelet(x, sigma=1.0, w=5.0):
    return np.exp(-x**2 / (2 * sigma**2)) * np.cos(w * x)

# Generate x values
x_values = np.linspace(-5, 5, 500)

# Generate y values using the Morlet wavelet function
y_values = morlet_wavelet(x_values, 1, 6)




# Plot the wavelet
plt.figure(figsize=(6, 3))

plt.plot(x_values, y_values, linewidth = 4, color = "#b194e3")
# plt.plot(x_values + 0.20, y_values, linewidth = 4, color = "#835bc9")

plt.xlim(-4, 4)

# plt.title("Morlet Wavelet")
# plt.xlabel("x")
# plt.ylabel("Amplitude")

plt.axis("off")

# Save the plot as an SVG file
plt.savefig("morlet_wavelet.svg", format="svg", transparent=True)

# Show the plot (optional)
plt.show()
```
